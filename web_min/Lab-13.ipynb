### Defined the function to calculate the difference between 2 times in minutes

```python
def convert_diff_to_min(t1, t2):
    
    u1 = int(t1.split(":")[0])
    v1 = int(t1.split(":")[1])
    
    u2 = int(t2.split(":")[0])
    v2 = int(t2.split(":")[1])
    
    if(u1==u2):
        time_in_min = abs(v2-v1)
    else:
        time_in_min = abs(60-max(v1,v2))+60*abs(u2-u1-1)+min(v1,v2)
    
    return(time_in_min)

convert_diff_to_min("01:33:00","03:10:00")
```

### Sessionization - h1 heuristics 

```python
theta = 30 # let say, theta for h1 heuristic is 30 Minutes
```

```python
start_time = df.iloc[0,0]
print(df.iloc[2,0])
convert_diff_to_min(str(start_time),str(df.iloc[2,0]))
```

```python
def sessionization_h1(df,user_id):
    j=1
    start_time = df.iloc[0,0]
    session_df = pd.DataFrame()
    for i in range(df.shape[0]):
        if(convert_diff_to_min(str(start_time),str(df.iloc[i,0]))<=theta):
            session_df = session_df.append(df.iloc[i:i+1,:])
        else:
            print("Session ",j," for User-id ",user_id,": -")
            print(session_df)
            print()
            start_time = df.iloc[i,0]
            session_df = pd.DataFrame()
            session_df = session_df.append(df.iloc[i:i+1,:])
            j=j+1
    print("Session ",j," for User-id ",user_id,": -")
    print(session_df)
    print()
```
